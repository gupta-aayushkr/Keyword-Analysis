```python
import pandas as pd
df = pd.read_csv("Discord.csv")
df2 = pd.read_csv("Discord.csv")
```

```python
df2["Sum"] = round(df.iloc[:, -12:].sum(axis=1),2)
df2["6-Months"] = round(df.iloc[:, -6:].sum(axis=1) - df.iloc[:, -12:-6].sum(axis=1),2)
df2["3-Months"] = round(df.iloc[:, -3:].sum(axis=1) - df.iloc[:, -6:-3].sum(axis=1),2)
df2["1-Month"] = round(df.iloc[:, -1:].sum(axis=1) - df.iloc[:, -2:-1].sum(axis=1),2)
df2 = df2.sort_values(by="Sum", ascending=False)
df2.head(10)
```

```python
df_dropped = df2.iloc[:, [1,15,16,17,18]]
df_dropped.head(10)
```
