````python
import streamlit as st
import pandas as pd
import numpy as np
from collections import Counter
import re
from nltk.util import ngrams
import nltk
import os
````

````python
df = pd.read_csv('CSV Files/MasterSheet.csv')
keywords = df['Month']
all_keywords = ' '.join(keywords)
print(all_keywords)
words = re.findall(r'\w+', all_keywords)
print(words)
````

````python
words = [word.lower() for word in words]
print(words)
````

````python
word_counts = Counter(words)
print(word_counts)
````

````python
total_words = sum(word_counts.values())

# Calculate the frequency as a percentage and sort
word_freq_percentage = {}
for word, count in word_counts.items():
    word_freq_percentage[word] = (count / total_words) * 100

print(word_freq_percentage)
````

````python
sorted_word_freq_percentage = dict(sorted(word_freq_percentage.items(), key=lambda item: item[1], reverse=True))
print(sorted_word_freq_percentage)
````

Certainly, let's break down the provided code line by line and explain how the `sorted()` function works with the `lambda` function:

```python
word_freq = {"apple": 5, "banana": 2, "cherry": 8, "date": 3, "fig": 1}

```

In this line, you've defined a dictionary named `word_freq` with keys representing words and values representing their frequencies.

```python
sorted_word_freq = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True))

```

1. `word_freq.items()` returns a view of the dictionary's (key, value) pairs. It's an iterable that looks like `dict_items([('apple', 5), ('banana', 2), ('cherry', 8), ('date', 3), ('fig', 1)])`.
2. The `sorted()` function takes this iterable of (key, value) pairs and sorts them based on the values (`item[1]`) using the `key` parameter.
3. **The `key` parameter specifies how the sorting should be done. Here, we're using a `lambda` function: `lambda item: item[1]`. This lambda function takes an (key, value) pair and returns the value (`item[1]`) which will be used for sorting. In other words, it's saying "sort based on the values of the dictionary".**
4. The `reverse=True` parameter tells the `sorted()` function to sort in descending order.
5. The result of the `sorted()` function is a sorted list of (key, value) pairs. The highest values (frequencies) come first due to the `reverse=True`.
6. Finally, the `dict()` constructor is used to convert the sorted list of (key, value) pairs back into a dictionary. This dictionary is assigned to the variable `sorted_word_freq`.

```python
print(sorted_word_freq)  # Output: {'cherry': 8, 'apple': 5, 'date': 3, 'banana': 2, 'fig': 1}

```

This line prints the sorted dictionary, where the words are sorted based on their frequencies in descending order. The output shows the words with their corresponding frequencies.

In summary, the provided code sorts a dictionary by its values (frequencies) in descending order using the `sorted()` function and a `lambda` function as the sorting key. It then converts the sorted list of (key, value) pairs back into a dictionary.

````python
sorted_word_freq_percentage = {}
sorted_items = sorted(word_freq_percentage.items(), key=lambda item: item[1], reverse=True)
for key, value in sorted_items:
    sorted_word_freq_percentage[key] = value
print(sorted_word_freq_percentage)
````
